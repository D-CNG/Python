<h1>Singapore Crude Birth Rate 2010 onwards</h1>

```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
import numpy
with open("crude-birth-rate.csv") as birthRate:
    birthR = csv.reader(birthRate,delimiter=",")
    birthL = [] #data list
    for column in birthR:
        birthL.append([column[0],column[2]]) #filling the list
        
    #removing data before year 2010
    birthL.remove(birthL[0]) #removing header
    dataY = int(birthL[0][0]) #year the data starts
    startY = 2010 #year of data we want
    diffY = startY - dataY 
    birthL = birthL[diffY:]
    
    #lists for respective axis
    bYear = []
    bRate = []
    
    #filling list with respective data
    for b in birthL:
        bYear.append(b[0])
        bRate.append(float(b[1])/10)
    
    plt.rcParams["figure.figsize"] = (10,6)#changing graph size
    
    #plot graphs
    plt.plot(bYear, bRate)
    plt.bar(bYear, bRate, color ='maroon', width = 0.4)
    plt.title('Birth Rate Vs Year')
    plt.xlabel('Year')
    plt.ylabel('Birth Rate')
    plt.show()
```

<h1>Total Female Residents</h1>

```python
with open("gender-ratio.csv") as genderRatio:
    genderR = csv.reader(genderRatio,delimiter=",")
    genderL = []
    dataY = 0
    for column in genderR:
        genderL.append([column[0],column[1],column[3]])
        if column[0] != 'year':
            if int(column[0])<2010:
                dataY += 1
                
    #removing data before year 2010  
    genderL.remove(genderL[0]) # remove list header
    genderL = genderL[dataY:] #remove data from unwanted years
    
    #graph axis lists
    tYear = []
    total = []
    y=0
    t=0
    #filling list with respective data
    for d in genderL:
        if d[1] == "Total Female Residents":
            if y != d[0]:
                if y != 0:
                    tYear.append(y)
                    total.append(t/1000000)
                y = d[0]
                t = 0
            if y == d[0]:
                t = t + int(d[2])
    plt.rcParams["figure.figsize"] = (10,5)#changing graph size         
        
    plt.plot(tYear, total)
    plt.bar(tYear, total, color ='maroon', width = 0.4)
    plt.title('Female Vs Year')
    plt.xlabel('Year')
    plt.ylabel('Female')
    plt.show()
```

```python
# plot lines
plt.rcParams["figure.figsize"] = (10,5)
#plt.plot(bYear, bRate,label = "birth rate")
plt.plot(bYear, bRate,label = "birth rate")
plt.plot(tYear, total,label = "population")
plt.legend()
plt.show()
```

<h1> Brides education qualification </h1>

```python
with open("brides-education.csv") as brideducation:
    brideEdu = csv.reader(brideducation,delimiter=",")
    bride_Edu_L = []
    dataY = 0
    for column in brideEdu:
        bride_Edu_L.append([column[0],column[1],column[2],column[3]])
        if column[0] != 'year':
            if int(column[0])<2010:
                dataY += 1
                
    bride_Edu_L.remove(bride_Edu_L[0])#remove list header
    bride_Edu_L = bride_Edu_L[dataY:]#remove data for unwanted years
    
    #list for each respective education level 
    blevel_3 = []
    blevel_2 = []
    blevel_1 = []
    lYear = []
    year = 0
    
    #filling list with respective data
    for b in bride_Edu_L:
        if b[1]=='First Marriages For Brides':
            if b[0] != year:
                if b[0] != 0:
                    lYear.append(b[0])
                    year = b[0]
            if b[2] == 'University':
                blevel_3.append(b[3])
            if b[2] == 'Post-secondary':
                blevel_2.append(b[3])
            if b[2] == 'Secondary & Below':
                blevel_1.append(b[3])
                
    plt.rcParams["figure.figsize"] = (8,8)#changing graph size           
    #plot graph lines
    plt.plot(lYear, blevel_1, label = 'Secondary & Below', marker='o')
    plt.plot(lYear, blevel_2, label = 'Post-secondary', marker='o')
    plt.plot(lYear, blevel_3, label = 'University', marker='o')
    
    plt.legend()
    plt.show()
```

<h1> Education of mother </h1>

```python
with open("birth-rate-mother-education.csv") as birtheducation:
    birthEdu = csv.reader(birtheducation,delimiter=",")
    br_m_list = []
    s = 0
    for column in birthEdu:
        br_m_list.append([column[0],column[2],column[4]])
        if column[1]=='OTHERS' :
            s += 1
            
    br_m_list.remove(br_m_list[0])
    br_m_list = br_m_list[:s]
    
    #graph list for each education level
    mlevel_4 = []
    mlevel_3 = []
    mlevel_2 = []
    mlevel_1 = []
    mlYear = []
    
    #variables used to find total data for that year
    level1 = 0
    level2 = 0
    level3 = 0
    level4 = 0
    year = 0 
    
    plt.rcParams["figure.figsize"] = (30,10)#changing graph size
    
    #filling list with respective data
    for l in br_m_list:
        if year != l[0]:
            if year != 0:
                mlYear.append(year)
                mlevel_1.append(level1)
                mlevel_2.append(level2)
                mlevel_3.append(level3)
                mlevel_4.append(level4)
            #reset for the next set of years
            year = l[0]
            level1 = 0
            level2 = 0
            level3 = 0
            level4  = 0
           
        #if the year is the same add the data together
        if l[1] == '"A" LEVEL/DIPLOMA':
            level4 += int(l[2])
        if l[1] == '"N" LEVEL/"O" LEVEL':
            level3 += int(l[2])
        if l[1] == 'PSLE':
            level2 += int(l[2])
        if l[1] == 'NO QUALIFICATION':
            level1 += int(l[2])
    
    
    #plot graph lines
    plt.plot(mlYear, mlevel_4, label = '"A" LEVEL/DIPLOMA', marker='o')
    plt.plot(mlYear, mlevel_3, label = '"N" LEVEL/"O" LEVEL', marker='o')
    plt.plot(mlYear, mlevel_2, label = 'PSLE', marker='o')
    plt.legend()
    plt.show()
```
